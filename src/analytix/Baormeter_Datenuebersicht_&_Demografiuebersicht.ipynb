```python
# Chi²-Test August 2022
# Chi²-Test: Geschlecht vs Kinder im Haushalt (August 2022)
table1 = pd.crosstab(df_2022_august["Geschlecht"], df_2022_august["Kinder im Haushalt (<18)"])
chi2, p, _, _ = stats.chi2_contingency(table1)
print("Chi²-Test August 2022")
print("Geschlecht vs Kinder (Aug 2022): chi2 = {:.3f}, p = {:.3f}".format(chi2, p))
# Chi²-Test: Altersgruppe vs Kinder im Haushalt (August 2022)
table2 = pd.crosstab(df_2022_august["Altersgruppe"], df_2022_august["Kinder im Haushalt (<18)"])
chi2b, pb, _, _ = stats.chi2_contingency(table2)
print("Altersgruppe vs Kinder (Aug 2022): chi2 = {:.3f}, p = {:.3e}".format(chi2b, pb))

#Chi²-Test Oktober 2022
# Chi²-Test: Geschlecht vs Kinder im Haushalt (Oktober 2022)
table3 = pd.crosstab(df_2022_okt["Geschlecht"], df_2022_okt["Kinder im Haushalt (<18)"])
chi2c, pc, _, _ = stats.chi2_contingency(table3)
print("\nChi²-Test Oktober 2022")
print("Geschlecht vs Kinder (Okt 2022): chi2 = {:.3f}, p = {:.3f}".format(chi2c, pc))
# Chi²-Test: Altersgruppe vs Kinder im Haushalt (Oktober 2022)
table4 = pd.crosstab(df_2022_okt["Altersgruppe"], df_2022_okt["Kinder im Haushalt (<18)"])
chi2d, pd_, _, _ = stats.chi2_contingency(table4)
print("Altersgruppe vs Kinder (Okt 2022): chi2 = {:.3f}, p = {:.3e}".format(chi2d, pd_))


# Chi²-Test April 2023
# Chi²-Test: Geschlecht vs Kinder im Haushalt (Oktober 2022)
table3 = pd.crosstab(df_2023_april["Geschlecht"], df_2023_april["Kinder im Haushalt (<18)"])
chi2c, pc, _, _ = stats.chi2_contingency(table3)
print("\nChi²-Test April 2023")
print("Geschlecht vs Kinder (April 2023): chi2 = {:.3f}, p = {:.3f}".format(chi2c, pc))
# Chi²-Test: Altersgruppe vs Kinder im Haushalt (Oktober 2022)
table4 = pd.crosstab(df_2023_april["Altersgruppe"], df_2023_april["Kinder im Haushalt (<18)"])
chi2d, pd_, _, _ = stats.chi2_contingency(table4)
print("Altersgruppe vs Kinder (April 2023): chi2 = {:.3f}, p = {:.3e}".format(chi2d, pd_))
```

```python
# Haushaltsnettoeinkommens (April 2023)
# Definierte Reihenfolge der Einkommenskategorien
income_order = ['<1000€', '1000€ bis 2000€', '2000€ bis 3000€',
                '3000€ bis 4000€', '4000€ bis 5000€', '>5000€']

# Kategorische Sortierung hinzufügen
df_2023_april["Einkommensklasse_cat"] = pd.Categorical(
    df_2023_april["Haushaltsnettoeinkommen"],
    categories=income_order,
    ordered=True
)

# Häufigkeiten je Kategorie, sortiert
einkommens_counts = df_2023_april["Einkommensklasse_cat"].value_counts().sort_index()

# Werte und Labels
values = einkommens_counts.values
labels = einkommens_counts.index

# Colormap und Normierung
cmap = plt.get_cmap('Blues')
norm = (values - values.min()) / (values.max() - values.min())
colors = cmap(norm)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, values, color=colors, edgecolor='black')

plt.xticks(rotation=45, ha='right')
plt.title("Verteilung des Haushaltsnettoeinkommens (April 2023)")
plt.xlabel("Nettoeinkommen")
plt.ylabel("Anzahl Befragte")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Ausgabe der Häufigkeiten
print("Anzahl Befragte pro Einkommenskategorie:")
for kategorie, anzahl in einkommens_counts.items():
    print(f"{kategorie}: {int(anzahl)} Befragte")
```
